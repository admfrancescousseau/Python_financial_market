1- # Importing libraries   

```python
import yfinance as yf   
import matplotlib.pyplot as plt
```

2- # Portfolio Stocks

```python
actions = ['PETR4.SA', 'VALE3.SA', 'BBAS3.SA', 'ITUB4.SA', 'AAPL', 'MSFT', 'AMZN', 'GOOG']
```

3- # Importing data from Financial Assets

```python
actions_data = yf.download(actions, start='2022-01-01')
```

4- # Creating the Chart

```python
actions_data['Adj Close'].plot()
plt.xlabel ('Data')
plt.ylabel ('Price')
plt.title ('Actions Tracker')
plt.legend (fontsize=5)
plt.show()
```

5- # Normalizing Prices

```python
data_normalized = actions_data ['Adj Close'] / actions_data ['Adj Close'].iloc[0]
data_normalized.plot()
plt.xlabel ('Data')
plt.ylabel ('Price')
plt.title ('Actions Tracker Normalized')
plt.legend (fontsize=5)
plt.show()
```
